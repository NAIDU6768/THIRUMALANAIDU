```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download necessary NLTK data
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize the lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Preprocessing function
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\d+', '', text)  # Remove digits
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    words = text.split()  # Split into words
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]  # Lemmatize and remove stopwords
    return ' '.join(words)  # Join words back into a string

# Apply preprocessing to the review_title and review_content columns
data['processed_review'] = data['review_title'] + " " + data['review_content']
data['processed_review'] = data['processed_review'].apply(preprocess_text)

# Display the processed reviews
data[['review_title', 'review_content', 'processed_review']].head()
```

```python
# Assuming 'rating' column exists for fake/real classification (using a threshold for rating to represent fake/real)
# You might need to adjust this threshold based on your data
# Convert the 'rating' column to numeric, handling errors by coercing to NaN
data['rating'] = pd.to_numeric(data['rating'], errors='coerce')
# Now you can apply the comparison and create the label
data['label'] = (data['rating'] < 3).astype(int)  # 1 for fake (low rating), 0 for real (high rating)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data['processed_review'], data['label'], test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize the TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000)

# Transform the text data into numerical vectors
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

# Initialize the Logistic Regression model
model = LogisticRegression()

# Train the model
model.fit(X_train_tfidf, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Predict the labels for the test set
y_pred = model.predict(X_test_tfidf)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print(classification_report(y_test, y_pred))
```

```python
# Example of a new review
new_review = "This product is awesome! Very durable and charges quickly. Highly recommend it."

# Preprocess the new review
processed_review = preprocess_text(new_review)

# Vectorize the new review
vectorized_review = tfidf_vectorizer.transform([processed_review])

# Predict the label (1 = fake, 0 = real)
prediction = model.predict(vectorized_review)

print("Original Review:", new_review)
print("Processed Review:", processed_review)
print("Prediction (1 = Fake, 0 = Real):", prediction[0])
```
